## Baseline 1: Multilingual Retrieval

In this baseline, we use a multilingual embedding model to map Arabic input sentences and English idioms into a shared semantic space.

The prediction is made by retrieving the most similar idiom based on cosine similarity.

This approach serves as a strong non-generative baseline for cross-lingual idiom prediction.

```python
import sys

!{sys.executable} -m pip install -q sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load model
model = SentenceTransformer("intfloat/multilingual-e5-base")

# Prepare corpus (idiom bank)
idiom_bank = df_train["target_text"].unique().tolist()

# Encode idioms (English)
idiom_embeddings = model.encode(
    ["passage: " + idiom for idiom in idiom_bank],
    normalize_embeddings=True,
    show_progress_bar=True
)
```

```python
def predict_idiom_e5(text, top_k=1):
    query_embedding = model.encode(
        ["query: " + text],
        normalize_embeddings=True
    )[0]
    
    scores = np.dot(idiom_embeddings, query_embedding)
    top_idx = np.argsort(scores)[::-1][:top_k]
    
    return [idiom_bank[i] for i in top_idx]
```

```python
def evaluate_e5(df, sample_size=2000):
    correct = 0
    
    sample_df = df.sample(n=sample_size, random_state=42)
    
    for _, row in sample_df.iterrows():
        pred = predict_idiom_e5(row["input_text"])[0]
        if pred == row["target_text"]:
            correct += 1
    
    accuracy = correct / sample_size
    return accuracy

e5_accuracy = evaluate_e5(df_test)
print("E5 Top-1 Accuracy:", e5_accuracy)
```

### Baseline Results

The multilingual retrieval model provides a strong baseline for the Arabic-to-English idiom mapping task.

This approach relies purely on semantic similarity and does not involve generative modeling.

Its performance reflects the quality of the shared embedding space and its ability to align Arabic context with English idiomatic expressions.

Because the Arabic dataset contains both literal and idiomatic realizations associated with the same canonical idiom, direct one-stage prediction introduces ambiguity. To reduce supervision noise, we first restrict the primary mapping task to idiomatic contexts, and later evaluate detection-based and retrieval-augmented extensions.

As an initial baseline, we restrict the Arabic context-to-idiom task to idiomatic examples only. This reduces label ambiguity caused by mixing literal and figurative realizations under the same canonical idiom.

The resulting subset is used to establish a cleaner benchmark before moving to a more advanced multi-stage framework that explicitly models literal versus idiomatic usage.

## Baseline 1: Generative Model (mT5)

In this baseline, we train a multilingual sequence-to-sequence model to directly generate the target idiom from an Arabic input sentence.

The model learns to map cross-lingual contextual meaning into a fixed idiomatic expression space.

This serves as the primary generative baseline for Task 3.

```python
from datasets import Dataset

def prepare_mt5_format(df):
    return Dataset.from_pandas(
        df[["input_text", "target_text"]].rename(
            columns={
                "input_text": "input",
                "target_text": "target"
            }
        )
    )

train_ds = prepare_mt5_format(idiom_train)
val_ds = prepare_mt5_format(idiom_val)
test_ds = prepare_mt5_format(idiom_test)
```

```python
from transformers import MT5Tokenizer

model_name = "google/mt5-small"  # start small, later we can scale
tokenizer = MT5Tokenizer.from_pretrained(model_name)

max_input_length = 128
max_target_length = 16

def preprocess(example):
    inputs = ["Arabic: " + text for text in example["input"]]
    
    model_inputs = tokenizer(
        inputs,
        max_length=max_input_length,
        truncation=True,
        padding="max_length"
    )
    
    labels = tokenizer(
        example["target"],
        max_length=max_target_length,
        truncation=True,
        padding="max_length"
    )
    
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_ds = train_ds.map(preprocess, batched=True)
val_ds = val_ds.map(preprocess, batched=True)
```

```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer

model_name = "google/mt5-small"

tokenizer = MT5Tokenizer.from_pretrained(model_name)
model = MT5ForConditionalGeneration.from_pretrained(model_name)

print("Model loaded successfully")
```

```python
# Step: format train / validation / test data for mT5
TASK_PREFIX = "Arabic to English idiom: "

df_train_mt5 = idiom_train.copy()
df_val_mt5 = idiom_val.copy()
df_test_mt5 = idiom_test.copy()

df_train_mt5["source_text"] = TASK_PREFIX + df_train_mt5["input_text"].astype(str)
df_val_mt5["source_text"]   = TASK_PREFIX + df_val_mt5["input_text"].astype(str)
df_test_mt5["source_text"]  = TASK_PREFIX + df_test_mt5["input_text"].astype(str)

df_train_mt5["target_text"] = df_train_mt5["target_text"].astype(str)
df_val_mt5["target_text"]   = df_val_mt5["target_text"].astype(str)
df_test_mt5["target_text"]  = df_test_mt5["target_text"].astype(str)

pd.set_option("display.max_colwidth", 300)

print("Formatted samples:")
display(df_train_mt5[["source_text", "target_text"]].head(5))
```

```python
# Step: tokenize source and target text for mT5

MAX_SOURCE_LEN = 128
MAX_TARGET_LEN = 16

def tokenize_function(examples_df, tokenizer, max_source_len=128, max_target_len=16):
    model_inputs = tokenizer(
        examples_df["source_text"].tolist(),
        max_length=max_source_len,
        truncation=True,
        padding="max_length"
    )

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples_df["target_text"].tolist(),
            max_length=max_target_len,
            truncation=True,
            padding="max_length"
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_encodings = tokenize_function(df_train_mt5, tokenizer, MAX_SOURCE_LEN, MAX_TARGET_LEN)
val_encodings   = tokenize_function(df_val_mt5, tokenizer, MAX_SOURCE_LEN, MAX_TARGET_LEN)
test_encodings  = tokenize_function(df_test_mt5, tokenizer, MAX_SOURCE_LEN, MAX_TARGET_LEN)

print("Tokenization completed.")
print("Train samples:", len(train_encodings["input_ids"]))
print("Validation samples:", len(val_encodings["input_ids"]))
print("Test samples:", len(test_encodings["input_ids"]))

print("\nExample tokenized input length:", len(train_encodings["input_ids"][0]))
print("Example tokenized label length:", len(train_encodings["labels"][0]))
```

```python
# Step: build PyTorch Dataset objects

import torch
from torch.utils.data import Dataset

class IdiomDataset(Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings["input_ids"])

    def __getitem__(self, idx):
        item = {
            key: torch.tensor(val[idx])
            for key, val in self.encodings.items()
        }
        return item

train_dataset = IdiomDataset(train_encodings)
val_dataset   = IdiomDataset(val_encodings)
test_dataset  = IdiomDataset(test_encodings)

print("Dataset objects created successfully.")
print("Train dataset size:", len(train_dataset))
print("Validation dataset size:", len(val_dataset))
print("Test dataset size:", len(test_dataset))

print("\nSample item keys:", train_dataset[0].keys())
print("input_ids shape:", train_dataset[0]["input_ids"].shape)
print("attention_mask shape:", train_dataset[0]["attention_mask"].shape)
print("labels shape:", train_dataset[0]["labels"].shape)
```

```python
# Step: define collator + training arguments (compatible version)

from transformers import DataCollatorForSeq2Seq, Seq2SeqTrainingArguments

data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model
)

training_args = Seq2SeqTrainingArguments(
    output_dir=str(PROJECT_ROOT / "deep_learning" / "models" / "arabic_context_to_idiom_mt5_baseline"),

    do_train=True,
    do_eval=True,

    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,

    num_train_epochs=3,
    learning_rate=5e-5,
    weight_decay=0.01,

    predict_with_generate=True,
    generation_max_length=16,

    fp16=torch.cuda.is_available(),

    logging_steps=100,
    save_total_limit=2,

    report_to="none"
)

print("Training arguments created successfully.")
print("Output dir:", training_args.output_dir)
print("FP16 enabled:", training_args.fp16)
```

```python
# Step: define evaluation metric (Exact Match)

import numpy as np

def compute_metrics(eval_preds):
    preds, labels = eval_preds

    # Decode predictions
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

    # Replace -100 in labels (ignore index) before decoding
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Clean text
    decoded_preds = [p.strip().lower() for p in decoded_preds]
    decoded_labels = [l.strip().lower() for l in decoded_labels]

    # Exact match
    correct = sum(p == l for p, l in zip(decoded_preds, decoded_labels))
    total = len(decoded_preds)

    accuracy = correct / total if total > 0 else 0

    return {
        "exact_match_accuracy": accuracy
    }

print("Metric function defined successfully.")
```

```python
# Step: initialize Trainer and start training

from transformers import Seq2SeqTrainer

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print("Trainer initialized. Starting training...\n")

trainer.train()
```
